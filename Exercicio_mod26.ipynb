```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 1 . Reduzindo correlações

```python
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
plt.scatter(df.p1,df.p2)
```

# a) Transformação linear da média e diferença dos valores

```python
df['média']=df.mean(axis=1)
df['diferença']=df.p2-df.p1
df.style.format(precision=1,decimal=',')
```

# b) Calculando de correlação entre os valores de p1 e p2

```python
df['p1'].corr(df['p2'])
```

# c) Calculando a correlação entre os valores transformados(média e diferença)

```python
df['média'].corr(df['diferença'])
```

É possível verificar que a correlação entre os dados transformados é muito pequena em comparação a p1 e p2. 

# 2. 

```python
df['soma']=df['p1']+df['p2']

df.style.format(precision=1,decimal=',')
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(df['p1'], bins=10)
plt.title('Distribuição da p1')


plt.subplot(1, 2, 2)
plt.hist(df['soma'], bins=10)
plt.title('Distribuição da soma')

plt.show()
```

- É possível verificar que após a transformação a distribuição ficou mais simétrica
