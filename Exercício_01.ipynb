```python
#from pandas_datareader import data as web
import plotly.express as px
from plotly import graph_objects
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()

```

```python
df = pdr.get_data_yahoo(["petr4.sa",'vale3.sa','goll4.sa'], start="2020-12-29")
df2=pd.DataFrame(df)
df2
```

####1. Amplitude de variação dos ativos



```python

amplitude=df2['Adj Close'].diff().stack().reset_index()
```

####2. Gráfico para visualização dos dados

```python

amplitude.columns=['DATA','ATIVOS','VALOR']
fig=px.line(amplitude,x='DATA',y='VALOR',color='ATIVOS',template='simple_white')
fig.show()
```

É possível visualizar que a **VALE** teve maior volatividade em comparação ao outros ativos
