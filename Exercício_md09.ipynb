O **Castanheiras** foi o que teve menor número de nascimentos com 32 registros.

### Tarefa 04

```python
nasci_marco=df.loc[(df['DTNASC']>='2019-03-01') & (df['DTNASC']<='2019-03-31')]
nasci_marco['munResNome'].value_counts()
porto_velho=nasci_marco[nasci_marco['munResNome']== 'Porto Velho']
```

```python
porto_velho.groupby(['munResNome'])['QTDFILVIVO'].max()
#Máximo de filhos vivos
```

```python
porto_velho.groupby(['munResNome'])['QTDFILVIVO'].min()
#Mínima de filhos vivos
```

```python
porto_velho.groupby(['munResNome'])['IDADEPAI'].mean()
#Idade média dos pais 
```

```python
porto_velho.groupby(['munResNome'])['IDADEPAI'].max()
#Idade máximo dos pais
```
