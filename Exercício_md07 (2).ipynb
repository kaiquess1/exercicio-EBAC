```python
categoria=aplication.select_dtypes(include='object')
nova_categoria=pd.get_dummies(categoria,drop_first=True)
aplication.drop(categoria.columns,axis=1)
novo_aplication=pd.concat([aplication,nova_categoria,credito])
novo_aplication.nunique()
```
