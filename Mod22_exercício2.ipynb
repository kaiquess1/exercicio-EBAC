```
#from pandas_datareader import data as web
import plotly.express as px
from plotly import graph_objects
import pandas as pd
import numpy as np
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()
```

* ## Carregando os dados

```
df = pdr.get_data_yahoo(["petr4.sa",'vale3.sa','goll4.sa'], start="2020-12-29")
df2=pd.DataFrame(df)
df2
```

* ## Média móvel

#### 1. Faça uma **média** móvel em 90 dias dos valores do fechamento

```
#Preciso fazer a media movel de 90 dias 
fechamento=df2['Adj Close'].stack().reset_index()
#Usando a função STACK() para empilhar os ativos
fechamento.columns=['Data','Ativos','Valor']
mm = 90
fechamento['media']=fechamento['Valor'].rolling(mm).mean()

```

#### 2. Faça uma **desvio** padrão móvel em 90 dias dos valores do fechamento

```

fechamento['desvio']=fechamento['Valor'].rolling(mm).std()
```

```
fechamento
```

 #### 3. Faça um gráfico de linhas com a média móvel

```
fechamento.dropna(inplace=True)
media=fechamento[['Data','Valor','Ativos','media']]
desvio=fechamento[['Data','Valor','Ativos','desvio']]
```

```
fig=px.line(media,x='Data',y='Valor',color='Ativos',template='simple_white')
fig.show()
```

### 4. Faça um gráfico com o desvio padrão

```
fig=px.line(desvio,x='Data',y='Valor',color='Ativos',template='simple_white')
fig.show()
```

### 5. Compare os dois gráficos para verificar se existe semelhança entre eles

* É possivel verificar que existe uma semelhança grande  entre os gráficos no periodo de maior volatividade dos três ativos. 

* ## Gráfico candlestick

```
indicadores=['Adj Close','Close','High','Low','Open','Volume']
ativo=['GOLL4.SA']*len(indicadores)
colunas=list(zip(indicadores,ativo))
```

```
gol=df2.loc['10/26/2022':,colunas]
gol.columns=indicadores
gol.head()
```

```
graph={
    'x':gol.index,
    'open':gol.Open,
    'close':gol.Close,
    'high':gol.High,
    'low':gol.Low,
    'type':'candlestick',
    'name':'GOLL4.SA',
    'showlegend':True
}
```

```
#from ipywidgets import Layout
#from sys import path_importer_cache
from plotly import graph_objects
from plotly.graph_objs import Layout
layout= Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title = 'Índice Gol',
    layout = layout
)

grafico.update_xaxes(showgrid=True,gridwidth=1,gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True,gridwidth=1,gridcolor='LightGrey')
grafico
```
